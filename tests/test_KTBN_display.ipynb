```python
import pyagrum as gum
import pyagrum.lib.notebook as gnb
import sys
import os
sys.path.insert(0, os.path.abspath(os.path.join(os.path.curdir, '..', 'lib')))
from KTBN_notebook import show_ktbn, show_unroll, show_time_slice
from KTBN import KTBN
```

```python
# Creation of a KTBN
ktbn = KTBN(k=2, delimiter='#')

# Ajouter des variables
ktbn.addVariable("A", temporal=True)
ktbn.addVariable("B", temporal=True)
ktbn.addVariable("C", temporal=False)

# Ajouter des arcs
ktbn.addArc(("A", 0), ("B", 0))
ktbn.addArc(("A", 0), ("A", 1))
ktbn.addArc(("C", -1), ("B", 1))
```

```python
# Display the KTBN
show_ktbn(ktbn)
```

```python
# Display the KTBN unrolled for 2 more time-slices
show_unroll(ktbn, 2)
```
